```python
#Control flow Statement    
#conditional statement if, if else,if elif else
#loop statement for loop,while loop 
#jumping statement break,continue,pass
```

```python
#conditional statement
age=18
if age>18:
    print("You are eligible for voting")
elif age==18:
    print("YES, You are eligible for voting")
elif age<18:
    print("Not eligible for voting")
else:
    print("Not eligible for voting")
```

```python
#looping statement for loop and while loop
list1=[1,2,3,4,5]
for x in list1:
    print(x)
for x in range(4,10):#parameter start 0,stop n-1,step 1
    print(x)
#while loop
i=1
while (i<=10):
    print(i)
    i+=1
```

```python
#create a program where we enter number and the program will add all number one by one and print until the user will enter q to quit the
sum = 0
while True:
    number = input("Enter a number (press q to quit): ")
    if number == 'q':
        print("Final sum:", sum)
        break
    sum += float(number)
    print("Current sum:", sum)
```

```python
#jumping statement
for i in range(1,11):
    if i==5:
        continue
    print(i)
```
